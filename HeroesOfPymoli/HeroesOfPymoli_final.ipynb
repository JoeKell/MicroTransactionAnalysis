## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueItems = len(purchase_data["Item ID"].unique())        
AveragePurchasePrice = purchase_data["Price"].mean()
TotalPurchases = purchase_data["Purchase ID"].count()
TotalRev = purchase_data["Price"].sum()

Purchasing_Summary=pd.DataFrame({
    "Number of Unique Items": [uniqueItems],
    "Average Purchasing Price":[AveragePurchasePrice],
    "Total Number of Purchases":[TotalPurchases],
    "Total Revenue":[TotalRev]
})

Clean_Purchasing_Summary=Purchasing_Summary
Clean_Purchasing_Summary["Average Purchasing Price"]=Clean_Purchasing_Summary["Average Purchasing Price"].map("${:.2f}".format)
Clean_Purchasing_Summary["Total Revenue"]=Clean_Purchasing_Summary["Total Revenue"].map("${:,.2f}".format)

Clean_Purchasing_Summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
Age_Bin=purchase_data

# Create the bins in which Data will be held
bins = [0, 9, 14, 19, 24, 29, 34, 39, 499]

# Create the names for the five bins
group_names = ["<=9", "10-14", "15-19", "20-24", "25-29","30-34","35-39",">=40"]

Age_Bin["Age Range"] = pd.cut(Age_Bin["Age"], bins, labels=group_names, include_lowest=True)
Age_grp=Age_Bin.groupby(["Age Range"])

Age_Purch=Age_grp["Purchase ID"].count().reset_index()
Age_Identity_Summary=Age_grp["SN"].nunique().reset_index()
Age_Identity_Summary["Percentage"]=(Age_Identity_Summary["SN"]/Age_Identity_Summary["SN"].sum()*100).map("{:.2f}%".format)
Age_Identity_Summary=Age_Identity_Summary.rename(columns={"SN": "Total"})
Age_Identity_Summary



Age_Identity_Summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#Just changing the values from above
Age_Purch=Age_grp["Purchase ID"].count().reset_index()
Age_Purch=pd.merge(Age_Purch,Age_grp["Price"].mean().reset_index(name="Average Purchase Price"),on="Age Range")
Age_Purch=pd.merge(Age_Purch,Age_grp["Price"].sum().reset_index(name="Total Purchase Value"),on="Age Range")
Age_Purch=pd.merge(Age_Purch,purchase_data.groupby(["Age Range","SN"])["Price"].sum().reset_index(level=0,name="AVG Total Purchase Per Person").reset_index(drop=True).groupby(["Age Range"]).mean(),on="Age Range")

#Cleaning up formatting
Clean_Age_Purch=Age_Purch
Clean_Age_Purch["Average Purchase Price"]=Clean_Age_Purch["Average Purchase Price"].map("${:,.2f}".format)
Clean_Age_Purch["Total Purchase Value"]=Clean_Age_Purch["Total Purchase Value"].map("${:,.2f}".format)
Clean_Age_Purch["AVG Total Purchase Per Person"]=Clean_Age_Purch["AVG Total Purchase Per Person"].map("${:,.2f}".format)

Clean_Age_Purch
```
